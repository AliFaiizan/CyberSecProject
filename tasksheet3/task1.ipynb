```python
    
from glob import glob
from utils import load_data


train_files = sorted(glob("../datasets/hai-22.04/train1.csv"))
test_files  = sorted(glob("../datasets/hai-22.04/test1.csv"))

X, y = load_data(train_files, test_files)   # X = readings, y = attack labels
print(f"[INFO] Loaded dataset: {X.shape}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)

X_normal = X[y == 0]
X_train, X_val = train_test_split(
    X_normal, test_size=0.2, random_state=42, shuffle=True
)
y_train = y_val = None
```

```python
from torch.utils.data import DataLoader, TensorDataset
import torch
```

```python
train_loader = DataLoader(
            TensorDataset(torch.from_numpy(X_train).float()),
            batch_size=8,
            shuffle=True,
        )
val_loader = DataLoader(
            TensorDataset(torch.from_numpy(X_val).float()),
            batch_size=8,
            shuffle=False,
        )
```

```python

from task1 import VAE

input_dim = X.shape[1]
model = VAE(
        input_dim=input_dim,
        latent_dim=8,
        layer_type='dense',
        activation='relu',
        num_classes=None
    )
```

```python

from task1 import train_vae_reconstruction

device = "cuda" if torch.cuda.is_available() else "cpu"
train_vae_reconstruction(
            model, train_loader, val_loader,
            device=device, epochs=10
        )
```
